```python
#soal nomor 1 bagian A
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter
lambda_param = 8  # Rata-rata kedatangan pelanggan per jam
jam_per_hari = 12
hari = 30

# Simulasi kedatangan pelanggan
total_jam = jam_per_hari * hari
jumlah_kedatangan = np.random.poisson(lambda_param, total_jam)

# Visualisasi histogram
plt.figure(figsize=(10, 6))
plt.hist(jumlah_kedatangan, bins=range(0, max(jumlah_kedatangan) + 1), alpha=0.7, color='green', edgecolor='black')
plt.title('Histogram Jumlah Kedatangan Pelanggan')
plt.xlabel('Jumlah Pelanggan')
plt.ylabel('Frekuensi')
plt.xticks(range(0, max(jumlah_kedatangan) + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()

# Menghitung probabilitas tidak ada pelanggan dalam 1 jam
prob_tidak_ada = poisson.pmf(0, lambda_param)

# Hasil probabilitas
print(f"Probabilitas tidak ada pelanggan dalam 1 jam: {prob_tidak_ada:.4f}")
```

```python
#soal nomor 1 bagian C
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter
lambda_param = 8  # Rata-rata kedatangan pelanggan per jam
jam_per_hari = 12
hari = 30

# Simulasi kedatangan pelanggan
total_jam = jam_per_hari * hari
jumlah_kedatangan = np.random.poisson(lambda_param, total_jam)

# Visualisasi histogram
plt.figure(figsize=(10, 6))
plt.hist(jumlah_kedatangan, bins=range(0, max(jumlah_kedatangan) + 1), alpha=0.7, color='orange', edgecolor='black')
plt.title('Histogram Jumlah Kedatangan Pelanggan')
plt.xlabel('Jumlah Pelanggan')
plt.ylabel('Frekuensi')
plt.xticks(range(0, max(jumlah_kedatangan) + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()

# Menghitung probabilitas
# a. Probabilitas tidak ada pelanggan dalam 1 jam
prob_tidak_ada = poisson.pmf(0, lambda_param)

# b. Probabilitas lebih dari 10 pelanggan dalam satu jam
prob_lebih_dari_10 = 1 - poisson.cdf(10, lambda_param)

# c. Membandingkan hasil simulasi dengan distribusi teoritis Poisson
frekuensi_simulasi = np.bincount(jumlah_kedatangan)
frekuensi_teoritis = [poisson.pmf(k, lambda_param) * total_jam for k in range(len(frekuensi_simulasi))]

# Visualisasi perbandingan
plt.figure(figsize=(10, 6))
plt.bar(range(len(frekuensi_simulasi)), frekuensi_simulasi, alpha=0.7, label='Simulasi', color='purple', edgecolor='grey')
plt.plot(frekuensi_teoritis, label='Teoritis', color='pink', marker='o')
plt.title('Perbandingan Simulasi dan Distribusi Teoritis Poisson')
plt.xlabel('Jumlah Pelanggan')
plt.ylabel('Frekuensi')
plt.xticks(range(len(frekuensi_simulasi)))
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()

# Hasil probabilitas
print(f"Probabilitas tidak ada pelanggan dalam 1 jam: {prob_tidak_ada:.4f}")
print(f"Probabilitas lebih dari 10 pelanggan dalam satu jam: {prob_lebih_dari_10:.4f}")

```
